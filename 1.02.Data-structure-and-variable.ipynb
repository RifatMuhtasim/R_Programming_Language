# Matrix

```r
# We can create a Matrix using the matrix() function.
# If we don't provide number of column. R will create a matrix 1 column and multiple rows.

mat_var <- matrix(1:6)
print(mat_var)
```

```r
# You can  specify the number of rows and columns for the matrix 
mat_var <- matrix(1:6, nrow=3, ncol=2)
print(mat_var)
```

```r
# Matrices are constructed column wise. If we want to construct a matrix
# row wise, set the argument byrow = TRUE. 

mat_var = matrix(1:6, nrow=3, ncol=2, byrow=TRUE)
print(mat_var)
```

```r
typeof(mat_var)
length(mat_var)
class(mat_var)
attributes(mat_var)
```

```r
# To add rows or columns to a matrix use the rbind() and cbind() functions.

mat_var2 <- rbind(mat_var, c(7,8))
print(mat_var2)
mat_var3 <- cbind(mat_var2, c(11, 12, 13, 14))
print(mat_var3)
```

```r
# To access elements of a matrix, declare the index of the row and the column inside a []. 

mat_var3[2, 2] #Element in the 2nd row and the 2nd column
mat_var3[c(3, 4), 2]
```

```r
# If a field inside the bracket is empty, R will select all those elements.
mat_var3[ , 3]
```

```r
# To exclude columns or rows, use a negative index inside []
print(mat_var3[-2, ])  # Here it removes the first column
```

```r
# To assign a name to the rows and columns of a matrix, use the rownames() and colnames() functions.

rownames(mat_var3) <- c("row1", "row2", "row3", "row4")
colnames(mat_var3) <- c("column1", "column2", "column3")
mat_var3
```

```r
# Just like with vectors, we can replace specific elements of a matrix with an assignment statement. 

mat_var2[2,2] <- 10
print(mat_var2)
```

# Array

```r
# Creating an array

array1 <- array(c(1:27), dim=c(3,3,3), dimnames= list(c("r1", "r2", "r3"), c("c1", "c2", "c3"), c("m1", "m2", "m3") ))
print(array1)
```

```r
# Differently using dimnames() function

array2 <- array(c(1:18), dim=c(2,3,3))
print(array2)
```

```r
# Indexing array in R

array2[ , , 2] # First Matrix
array2[ 1, 3, 2] # 1st row, 3rd Column for 2rd Matrix
print( array1[c(1,2),c(2,3),c(1,2)] ) # first and second row, second and third column of first and second matrices
```

```r
array1[1:2,1:2,2] # 1st two rows and columns of 2nd matrix
print(array1[-1,-3,-2]) # remove first row, third column and second matrix of array1
```

```r
# first and third row, second and third column of first and second matrix of array1

print(array1[c(T,F,T), c(F,T,T), c(T,T,F)])

"Using Character Vectors"
print(array1[c("r1", "r3"), c("c2", "c3"), c("m1", "m2")])
```

```r
# Re-assignment to change specific values of an array

print(array1)
# Change the values of the 1st row of 1st matrix
array1[1, , 1] <- c(4,4,4)
"Array1 Value"
print(array1)
```

```r
# Arithmetic Operations of Array in R

test_array1 <- array(c(1:8), c(2,2,2))
test_array2 <- array(c(9:16), c(2,2,2))
"Array 1"
print(test_array1)
"Array 2"
print(test_array2)
```

```r
# Addition
array_addition = test_array1 + test_array2
print(array_addition)
```

```r
# Subtraction
array_subtraction = test_array1 - test_array2
print(array_subtraction)
```

```r
# Multiplication
array_multiplication = test_array1 * test_array2
print(array_multiplication)
```

```r
#Division
array_division = test_array1 / test_array2
print(array_division)
```

```r
# Taking Power
taking_power = test_array1 ^ 2
print(taking_power)
```

# Factors

```r
factor1 <- factor(x = c("Male", "Female", "Female", "Male", "Female", "Male"))
print(factor1)
typeof(factor1)
str(factor1)
```

```r
print(factor1[1])
factor1[6] <- "Female"
print(factor1)
```

```r
factor2 <- factor(x = c("Female", "Male", "Male", "Female", "Male", "Female"))
print(factor2)
''
factor3 <- factor(x = c("Dog", "Cat", "Apple", "Elephent", "Ball"))
print(factor3)
str(factor3)
```

```r
# Without defining levels, R orderes the levels alphabetically considering as nominal variable

factor4 <- factor(x = c("Male", "Female", "Female", "Male", "Female", "Male"), levels = c("Male", "Female"))
print(factor4)
typeof(factor4)
str(factor4)
```

```r
# Change the Specific Value

factor5<- factor(x = c("Male", "Female", "Female", "Male", "Female", "Male"), levels = c("Male", "Female", "Other"))
print(factor5)
''
factor5[6]<- "Other"
print(factor5)
str(factor5)
```

```r
# Ordinal Variables as Factor

temperature <- factor(x = c("High", "Low", "Low", "High", "Medium"), ordered=TRUE, levels = c("Low", "Medium", "High"))
print(temperature)
str(temperature)
```

```r
#Renaming Factor labels

factor6 <- factor(x= c("Male", "Female", "Female", "Male", "Female", "Male"), levels= c("Male", "Female"), labels= c("M", "F"))
print(factor6)
```

```r
factor7 <- factor(x= c(1,1,2,2,2,1))
print(factor7)
```

```r
factor8 <- factor(x= c(1,1,2,2,2,1), levels= c("Male", "Female"))
print(factor8)
''
factor9 <- factor(x= c(1,1,2,2,2,1), labels= c("Male", "Female"))
print(factor9)
```

```r
factor10 <- factor(x= c("Male", "Female", "Female", "Male", "Female", "Male"), levels= c("Male", "Female"), labels= c(1, 2))
print(factor10)
```

```r
# Convert binary factor levels to '0' and '1'

binary_gender <- c("Male", "Female")
as.factor(binary_gender)
contrasts(as.factor(binary_gender))
```

# Data Frames

```r
# We can create a data frame using the data.frame() function.

x <- c(2,5,3,6,1,6,2,3,1,5)
y <- c(54, 14, 72, 22, 52, 83, 12, 16, 93, 22)
df <- data.frame(x, y)
print(df)
```

```r
# Get the number of rows and Number of column

ncol(df)
nrow(df)
```

```r
# To add columns to a data frame use the cbind() function. 

df <- cbind(df, NewCol= c(TRUE, FALSE, FALSE, TRUE, FALSE, FALSE, TRUE, FALSE, FALSE, TRUE ))
df
```

```r
# To add rows to a data frame use the rbind() function. We must specify the number of columns

df <- rbind(df, data.frame(x= 6, y= 99, NewCol=TRUE))
df

class(df)
typeof(df)
```

```r
# We can also add columns to a data frame using the $. 

df$z <- c(10:20)
df
```

```r
# Use [[]] to access a column of a data frame by its position and $ to access it by its name. 

# Access the Column 1
df[[1]]
df[[1]][2]
df$z
df$z[1]
```

```r
# Just like a matrix, access elements by declaring the index of the row and the column inside a []. 

'All rows, column 1: '
df[, 1]  
'All rows, column 1 and 4:'
df[, c(1, 4)]
'Row 2, columns 1 and 3:'
df[2, c(1, 3)]
'Row 2, all columns:'
df[2, ]
```

# 2-Way Frequency Table

```r
A <- factor(x = c("Male", "Female", "Female", "Male", "Female", "Male"), levels = c("Male", "Female"))
B <- factor(x = c("A", "A", "B", "A", "B", "B"), levels = c("A", "B"))

mytable <- table(A, B)
print(mytable)
```

```r
'A frequencies (summed over B):'
margin.table(mytable, 1)

'B frequencies (summed over A):'
margin.table(mytable, 2)

'cell percentages:'
prop.table(mytable)

'row percentage:'
prop.table(mytable, 1)

'column percentage:'
prop.table(mytable, 2)
```
